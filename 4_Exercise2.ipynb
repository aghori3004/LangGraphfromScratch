<h1> Exercise 2: Calculator Agent </h1>

### The Challenge
Build a 'Calculator' agent that can perform simple arithmetic based on user input.

**Requirements:**
1. Define a State that holds:
    - `intgrs`: A list of numbers.
    - `name`: A string.
    - `result`: A string (output).
    - `operator`: A string (either "+" or "*").
2. Create a node that checks the `operator`.
    - If `operator` is "+", sum the numbers.
    - If `operator` is "*", multiply them.
3. Ensure the result message greets the user and shows the answer.

### The Solution
Ready? Let's check the solution.

```python
from typing import TypedDict, List
from langgraph.graph import StateGraph
import numpy as np
```

### Step 1: Define State
This matches the requirements. We use `TypedDict` to enforce the structure.

```python
class AgentState(TypedDict):
    intgrs: List[int]
    name: str
    result: str
    operator: str
```

### Step 2: The Logic Node
Here we implement the arithmetic logic. We check the `operator` field and do the math.

```python
def condNode(state: AgentState) -> AgentState:
    """
    This is a conditional operation function for addition and multiplication.
    """
    if(state["operator"] == "+"):
        # Use sum() for addition
        state["result"] = f"Hi {state["name"]}, your answer is {sum(state["intgrs"])}."
    elif(state["operator"] == "*"):
        # Use np.prod() for multiplication
        state["result"] = f"Hi {state["name"]}, your answer is {np.prod(state["intgrs"])}."
    return state
```

### Step 3: Build & Run
Add the node, compile, and run it with some test data.

```python
graph = StateGraph(AgentState)
graph.add_node("condNode", condNode)
graph.set_entry_point("condNode")
graph.set_finish_point("condNode")

app = graph.compile()
```

```python
result = app.invoke({"name": "Jack Sparrow", "intgrs": [1,2,3,4], "operator": "*"})
print(result["result"])
```
